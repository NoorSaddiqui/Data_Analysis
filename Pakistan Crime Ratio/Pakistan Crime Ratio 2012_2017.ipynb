Creating HeatMaps of Data

```python
corelation = data.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(corelation, xticklabels = corelation.columns , yticklabels = corelation.columns  )
plt.show()
```

```python
corr= data.corr()
plt.figure(figsize =(5,5))

sns.heatmap(corr,vmax=.8, linewidths = 0.01, square = True, cmap = 'Greens', linecolor = 'black' )
plt.title('corelation between features')
plt.show()
```

```python
corr= data.corr()
plt.figure(figsize =(5,5))

sns.heatmap(corr,vmax=.8, linewidths = 0.01, square = True, cmap = 'Blues', linecolor = 'white' )
plt.title('corelation between features')
plt.show()
```
